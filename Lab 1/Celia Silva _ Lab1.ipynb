```python
# --- Treino (Versão com Link Alternativo) ---
import pandas as pd
import pickle as p1
from sklearn import linear_model
import numpy as np

# 1. Carregamento dos dados (Link alternativo do GitHub para fugir do erro 502 da UCI)
print("A carregar dados do link alternativo...")
# Este link contém o mesmo arquivo 'winequality-white.csv'
URL_ALT = 'https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/winequality-white.csv'

try:
    # Tentamos carregar do GitHub
    df = pd.read_csv(URL_ALT) # No GitHub desse usuário o separador já é vírgula

    # Se por acaso o separador for diferente, o pandas avisa,
    # mas o link acima é um padrão bem aceito.
    print("Dados carregados com sucesso!")
except Exception as e:
    print(f"Erro ao carregar: {e}")

# 2. Separação de Features (X) e Target (y)
X = df.drop(columns=['quality'])
y = df[['quality']]

# 3. Definição do tamanho dos dados
M = len(X)
m = int(2 * M / 3)

print(f"Nº total de dados: {M}")
print(f"Nº total de dados de treino: {m}")

# 4. Divisão Treino e Teste
data_X = X.iloc[:m]
data_Y = y.iloc[:m]
datateste_X = X.iloc[m:]
datateste_Y = y.iloc[m:]

# 5. Guardar dados de teste
Dadosteste = [datateste_X, datateste_Y]
with open("Dadosteste.pkl", "wb") as Dt:
    p1.dump(Dadosteste, Dt)

# 6. Treinar o Modelo
regr = linear_model.LinearRegression()
Modelo = regr.fit(data_X, data_Y)

# 7. Guardar o modelo treinado
with open('wineQpredictor', 'wb') as PickleModelo:
    p1.dump(Modelo, PickleModelo)

print("\nConcluído! 'wineQpredictor' e 'Dadosteste.pkl' foram gerados.")
```

```python
# --- CÓDIGO DE TREINO CORRIGIDO ---
import pandas as pd
import pickle as p1
from sklearn import linear_model
import numpy as np

# 1. CARREGAMENTO ALTERNATIVO (Ignora o fetch_ucirepo pois o servidor caiu)
print("A carregar dados do link alternativo...")
URL = 'https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/winequality-white.csv'

try:
    # O arquivo no GitHub usa vírgula como separador
    df = pd.read_csv(URL)

    # 2. DEFINIÇÃO DE FEATURES E TARGET (Igual ao que o fetch faria)
    # No CSV, a última coluna é 'quality'
    features = df.drop(columns=['quality'])
    targets = df[['quality']]

    M = len(features)
    print("nº total de dados:", M)
    print("nº total de parâmetros:", features.shape[1], " + 1")

    m = int(2*M/3)
    print("nº total de dados de treino:", m)

    # 3. SEPARAÇÃO DOS DADOS
    data_X = features.iloc[:m]
    data_Y = targets.iloc[:m]

    # Restantes dados de teste
    datateste_X = features.iloc[m:]
    datateste_Y = targets.iloc[m:]
    Dadosteste = [datateste_X, datateste_Y]

    # 4. SALVAR DADOS DE TESTE
    with open("Dadosteste.pkl", "wb") as Dt:
        p1.dump(Dadosteste, Dt)

    # 5. TREINO DO MODELO
    regr = linear_model.LinearRegression()
    Modelo = regr.fit(data_X, data_Y)

    # 6. SALVAR O MODELO
    with open('wineQpredictor', 'wb') as PickleModelo:
        p1.dump(Modelo, PickleModelo)

    print("wineQpredictor guardado com sucesso!")

except Exception as e:
    print(f"Erro ao carregar os dados: {e}")
```
